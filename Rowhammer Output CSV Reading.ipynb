```python
errors = {}
for i in range(labels.shape[0]):
    errors.update({i : (np.array(np.where(labels[i, :] >= 1))).flatten()})
```

```python
print(errors)
```

```python
pageCount = {}
for i in errors.keys():
    arr = errors.get(i)
    for j in arr:
        if j in pageCount.keys():
            #val = pageCount.get(j)
            pageCount[j] += 1
        else:
            pageCount.update({j:1})
print(pageCount)
```

```python
pageCountByVal = {k: v for k, v in sorted(pageCount.items(), key=lambda item: item[1])}
print(pageCountByVal)
pageCountByKey = {k: v for k, v in sorted(pageCount.items(), key=lambda item: item[0])}
print()
print(pageCountByKey)
            
```

```python
block26ErrorMap = {}
for i in range(block26errors.shape[0]):
    block26ErrorMap.update({i : (np.array(np.where(block26errors[i, :] >= 1))).flatten()})
```

```python
print(block26ErrorMap)
```

```python
block26pageCount = {}
for i in block26ErrorMap.keys():
    arr = block26ErrorMap.get(i)
    for j in arr:
        if j in block26pageCount.keys():
            #val = pageCount.get(j)
            block26pageCount[j] += 1
        else:
            block26pageCount.update({j:1})
print(block26pageCount)
```

```python
block26pageCountByVal = {k: v for k, v in sorted(block26pageCount.items(), key=lambda item: item[1])}
print(block26pageCountByVal)
block26pageCountByKey = {k: v for k, v in sorted(block26pageCount.items(), key=lambda item: item[0])}
print()
print(block26pageCountByKey)
```

```python
block26hundredCount = 0
block26twodredCount = 0
block26threedredCount = 0
block26fourdredCount = 0
block26fivedredCount = 0
for i in block26pageCount.values():
    if i >= 100:
        block26hundredCount += 1
    if i >= 200:
        block26twodredCount += 1
    if i >= 300: 
        block26threedredCount += 1
    if i >= 400:
        block26fourdredCount += 1
    if i >= 500:
        block26fivedredCount += 1
print(block26hundredCount)
print(block26twodredCount)
print(block26threedredCount)
print(block26fourdredCount)
print(block26fivedredCount)
```

```python
errorNums = np.zeros(len(block26ErrorMap.keys()))
for i in range(len(block26ErrorMap.keys())):
    errorNums[i] = block26ErrorMap[i].shape[0]
print(errorNums)
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(block26ErrorMap.keys(), errorNums)
plt.title("Number of Induced Flips by Hammering")
plt.xlabel("Hammered Page Index")
plt.ylabel("Number of Induced Flips")
plt.show()
```

```python
pointsList = []
for i in range(len(block26ErrorMap.keys())):
    for j in range(block26ErrorMap[i].shape[0]):
        pointsList.append((i, block26ErrorMap[i][j]))
#print(pointsList)
x,y = zip(*pointsList)
plt.scatter(x, y)
plt.title("Page with Induced Flip vs. Hammered Page")
plt.xlabel("Hammered Page")
plt.ylabel("Page with Induced Flip")
plt.show()
```
