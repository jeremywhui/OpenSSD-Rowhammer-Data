```python
errors = {}
for i in range(labels.shape[0]):
    errors.update({i : (np.array(np.where(labels[i, :] >= 1))).flatten()})
```

```python
print(errors)
```

```python
pageCount = {}
for i in errors.keys():
    arr = errors.get(i)
    for j in arr:
        if j in pageCount.keys():
            #val = pageCount.get(j)
            pageCount[j] += 1
        else:
            pageCount.update({j:1})
print(pageCount)
```

```python
pageCountByVal = {k: v for k, v in sorted(pageCount.items(), key=lambda item: item[1])}
print(pageCountByVal)
pageCountByKey = {k: v for k, v in sorted(pageCount.items(), key=lambda item: item[0])}
print()
print(pageCountByKey)
            
```
