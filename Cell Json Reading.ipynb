```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open ("cellLevel.json", "r") as file:
    data = json.load(file)
#print(data)
print(type(data))
print(data.keys())
print(data['errors'][0])
#print(data.values())
```

```python
pointsList = []
for i in range(len(data['errors'])):
    for j in data['errors'][i]:
        pointsList.append((i, j))
#print(pointsList)
```

```python
x,y = zip(*pointsList)
plt.figure(figsize = (20, 20))
plt.scatter(y, x)
plt.title("Cell with Flip vs. Number")
plt.ylabel("Number Index")
plt.xlabel("Cell with Flip")
plt.show()
```
